```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
```

```python
df = pd.read_json("..\\outputs\\11042022_1.json", orient='index')
```

```python
df.sample(10)
```

```python
df_ = pd.concat({'header': df['header'].apply(pd.Series),
                    '$STANDARD_INFORMATION': df['$STANDARD_INFORMATION'].apply(pd.Series),
                    '$FILE_NAME': df['$FILE_NAME'].apply(pd.Series),
                    '$DATA': df['$DATA'].apply(pd.Series),
                    '$INDEX_ROOT': df['$INDEX_ROOT'].apply(pd.Series),
                    '$INDEX_ALLOCATION': df['$INDEX_ALLOCATION'].apply(pd.Series)}, axis=1)
```

```python
df_.sample(10)
```

```python
df_.columns
```

```python
f, ax = plt.subplots(figsize=(13, 8))
sns.scatterplot(y=df_['header']['Entry number'],
                x=df_['$STANDARD_INFORMATION']['Creation time'],
                data=df_,
                ax=ax,
                x_bins=50)
```

```python
f, ax = plt.subplots(figsize=(13, 8))
sns.lineplot(y=df_['header']['Entry number'], x=df_['$STANDARD_INFORMATION']['Creation time'], data=df_)
```

```python
df1 = df_.loc[:, slice('$DATA', '$DATA')].rename(columns={'Run list': 'First cluster'}, level=1)
```

```python
df1 = df1.drop(columns=df1.columns.difference([('$DATA', 'First cluster')]))
```

```python
df1['$DATA'] = df1['$DATA'].loc[:, 'First cluster'].apply(lambda x: int(x[0][0]) if type(x) == list else x)
```

```python
df1
```

```python
df2 = pd.concat([df_, df1], axis=1)
```

```python
df2.sample(10)
```

```python
f, ax = plt.subplots(figsize=(13, 8))
sns.lineplot(y=df_['header']['Entry number'], x=df_['$DATA']['First cluster'], data=df_)
```

```python
f, ax = plt.subplots(figsize=(13, 8))
sns.scatterplot(y=df_['header']['Entry number'],
                x=df_['$DATA']['First cluster'],
                data=df_,
                ax=ax,
                x_bins=50)
```

```python
df3 = df2.droplevel(0, axis=1)
```

```python
df3 = df3[['$LogFile sequence number (LSN)', 'Sequence number', 'Allocation flag', 'Entry number', 'Path', 'Creation time', 'Filename', 'Run list', 'Data run', 'First cluster']]
df3.sample(10)
```

```python
df3['First cluster'] = df3['First cluster'].fillna(0).astype(int)
df3
```

```python
df3['Data run cluster'] = df3.loc[:, 'Data run'].apply(lambda x: x[0][0] if type(x) == list else x)
```

```python
df3
```

```python
fig = go.Figure([go.Scatter(x=df3['$LogFile sequence number (LSN)'], y=df3['Entry number'])])
fig.show()
```

```python
df3.plot(x='Entry number', y=['First cluster', 'Data run cluster'])
```
